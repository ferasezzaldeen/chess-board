```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class ChessBoard:
    def __init__(self):
        self.grid=np.zeros((8,8,3))
        self.red_queen=[]
        self.blue_queen=[]
        for col in range(len(self.grid)):
            for row in range(len(self.grid)):
                if col%2==0 and row%2==1:
                    self.grid[col,row]=[1,1,1]
                if col%2==1 and row%2==0:
                    self.grid[col,row]=[1,1,1]

    def add_red(self,row,col):
        red=[1,.2,0]
        self.grid[col,row]=red
        self.red_queen=[col,row]

    def add_blue(self,row,col):
        blue=[0,1,1]
        self.grid[col,row]=blue
        self.blue_queen=[col,row]

    def render(self):
        plt.imshow(self.grid)
    
    def is_under_attack(self):
        if self.blue_queen[0]==self.red_queen[0]:
            return True
        elif self.blue_queen[1]==self.red_queen[1]:
            return True
        elif abs(self.blue_queen[0]-self.red_queen[0])==abs(self.blue_queen[1]-self.red_queen[1]) :
            return True
        else:
            return False
```

```python
chess=ChessBoard()
chess.render()
```

```python
def test_virtical_attack():
    test=ChessBoard()
    test.add_blue(5,4)
    test.add_red(5,7)
    test.render()
    assert test.is_under_attack()==True
    print ('Passed')

test_virtical_attack()
```

```python
def test_horizental_attack():
    test=ChessBoard()
    test.add_blue(3,4)
    test.add_red(5,4)
    test.render()
    assert test.is_under_attack()==True
    print ('Passed')

test_horizental_attack()
```

```python
def test_diagonal_attack():
    test=ChessBoard()
    test.add_blue(2,5)
    test.add_red(4,7)
    test.render()
    assert test.is_under_attack()==True
    print ('Passed')
test_diagonal_attack()
```

```python
def test_not_under_attack():
    test=ChessBoard()
    test.add_blue(5,4)
    test.add_red(1,7)
    test.render()
    assert test.is_under_attack()==True
    print ('Passed')
test_not_under_attack()
```
